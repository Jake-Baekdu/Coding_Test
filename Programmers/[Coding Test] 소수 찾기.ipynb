# 문제 설명



```python
import timeit

def solution(n):
    start_time = timeit.default_timer()

    answer = [2]
    for x in range(3, n+1):
        for j in range(2, x//2+1):
            if x % j == 0:
                print("if문 안", x, j)
                break
        else :
            print("else 안", x)
            print()
            answer.append(x)
        print("else 밖", x)
        print()


    terminate_time = timeit.default_timer() # 종료 시간 체크  


    print("%f초 걸렸습니다." % (terminate_time - start_time)) 
    return len(answer)


solution(50)
```

print()를 사이에 찍어보니 에러가 뜨더라!

# 다른 풀이

소수 찾는 법을 공부하면서 에라토스테네스의 체를 이용하면 간단히 구현할 수 있다는 걸 알게 되었다.

밑에 코드와 함께 공부해보자.

https://ko.wikipedia.org/wiki/%EC%97%90%EB%9D%BC%ED%86%A0%EC%8A%A4%ED%85%8C%EB%84%A4%EC%8A%A4%EC%9D%98_%EC%B2%B4

```python
def solution(n):
    start_time = timeit.default_timer()

    # 짝수는 소수가 될 수 없으므로 n 미만의 자연수들 중 홀수들 만을 뽑아냄.
    answer = set([i for i in range(3, n+1, 2)])
    #answer 라는 배열에는 3부터 시작하는 n 미만의 자연수 배열이 들어갑니다.
    #이제 이 배열에서 배수들을 에라토스테네스의 체를 이용해서 모두 걸러냄
    for i in range(3, n//2+1, 2):
        if i in answer:
            answer -= set([i for i in range(i*2, n+1, i)])
    terminate_time = timeit.default_timer() # 종료 시간 체크  


    print("%f초 걸렸습니다." % (terminate_time - start_time)) 
    return len(answer) + 1

solution(100)
    
```

위에 코드는 set안에 list를 넣어서 조금 조잡해 보일 수도 있겠네요... 아래와 같이 깔끔히 구현할 수도 있습니다!

```python
def solution(n):
    num=set(range(2,n+1))

    for i in range(2,n//2+1):
        if i in num:
            num-=set(range(2*i,n+1,i))
    return len(num)
```

n//2+1까지 하는 이유는 어차피 n/2이 넘어가는 순간부터는 i*2가 의미가 없기 때문입니다.
