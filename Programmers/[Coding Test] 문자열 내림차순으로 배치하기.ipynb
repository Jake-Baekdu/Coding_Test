```python
def solution(s):
    return ''.join(sorted(s, reverse=True))

solution("ZbcCdAg")
```

## 후기

join을 문자열 사이에 추가하여 붙이는 것으로만 활용했었는데  ''빈 문자열에 넣으니 합쳐지는 것을 알았다. 


그리고 반복문을 써서 문자열이나 리스트를 뒤집는 법도 있지만 파라미터로 reverse를 넣는 법도 알게 되었다.






## 다른 풀이

```python
def solution(s):
    s = list(s)
    s.sort(reverse = True)
    print(s)
    answer = ""
    for i in s:
        answer = answer + i
    return answer

solution("ZbcCdAg")
```

배열에서만 sort를 쓸 수 있으니 배열을 추가한 것 같다.
