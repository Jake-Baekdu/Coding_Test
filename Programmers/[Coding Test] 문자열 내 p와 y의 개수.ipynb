# 문제 설명









출처 : https://programmers.co.kr/learn/courses/30/lessons/12916?language=python3


```python
def solution(s):
    
    lower_s = s.lower()
    print(lower_s)
    find_p = lower_s.count('p')
    find_s = lower_s.count('y')
    
    if find_p == find_s:
        return True
    if find_p == 0  and  find_s == 0:
        return True
    else : 
        return False
    
solution("pPoooyY")
```

## 후기

count에 대한 개념이 헷갈렸는데 다시 한번 공부할 수 있는 기회였다.






```python
from collections import Counter
def numPY(s):
    # 함수를 완성하세요
    c = Counter(s.lower())
    print(c)
    return c['y'] == c['p'] 



# 아래는 테스트로 출력해 보기 위한 코드입니다.
print( numPY("pPoooyY") )
print( numPY("Pyy") )
```

## 다른 풀이

보다 간단한 모듈로 문제를 푸셨다. Counter를 기억해두면 많은 비교를 할 때는 유용할 것 같다.



```python
def numPY(s):
    # 함수를 완성하세요
    return s.lower().count('p') == s.lower().count('y')

# 아래는 테스트로 출력해 보기 위한 코드입니다.
print( numPY("pPoooyY") )
print( numPY("ㅁㅁ") )
```

내가 푼 코드는 if문으로 제외를 두번이나 줬는데 한줄로 간단히 표현할 수 있다는 점에서 다시 한번 존경을 표한다.
