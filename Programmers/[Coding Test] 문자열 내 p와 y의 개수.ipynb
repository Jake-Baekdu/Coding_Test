```python
def solution(s):
    
    lower_s = s.lower()
    print(lower_s)
    find_p = lower_s.count('p')
    find_s = lower_s.count('y')
    
    if find_p == find_s:
        return True
    if find_p == 0  and  find_s == 0:
        return True
    else : 
        return False
    
solution("pPoooyY")
```

```python
from collections import Counter
def numPY(s):
    #  
    c = Counter(s.lower())
    print(c)
    return c['y'] == c['p'] 



#      .
print( numPY("pPoooyY") )
print( numPY("Pyy") )
```
