```python
def solution(w,h):
    if w == h:
        return w*h - w
    if w == 1 or h == 1:
        return 0
    if w != h:
        X, Y = w, h
        while(Y):
            X, Y = Y, X%Y
            print(X)
        return w*h - (w+h-X)
    
solution(8, 12)
```
