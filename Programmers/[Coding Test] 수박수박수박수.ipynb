# 문제 설명



```python
def solution(n):
    answer = ""
    for i in range(n):
        if i%2== 0:
            answer += "수"
        if i%2== 1:
            answer += "박"
    return answer

solution(4)
```

# 다른 풀이

```python
def water_melon(n):
    s = "수박" * n
    return s[:n]


# 실행을 위한 테스트코드입니다.
print("n이 3인 경우: " + water_melon(3));
print("n이 4인 경우: " + water_melon(4));
```

### 다른 컨프리헨션이나 슬라이딩과 같은 조금 파이썬같은 풀이가 간단한 것 같다.

```python
def water_melon(n):
    return "수박"*(n//2) + "수"*(n%2)


# 실행을 위한 테스트코드입니다.
print("n이 3인 경우: " + water_melon(3));
print("n이 4인 경우: " + water_melon(4));
```
