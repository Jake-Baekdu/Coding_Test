# 한글 번역

작은 개구리가 길 반대편으로 가고 싶어한다. 개구리는 현재 위치 X에 위치하며 Y보다 크거나 같은 위치에 도달하기를 원한다. 작은 개구리는 항상 일정한 거리인 D를 점프한다.

작은 개구리가 목표물에 도달하기 위해 수행해야 하는 최소 점프 수를 세어라.

Write a function:

def solution(X, Y, D)

X, Y, D의 세 정수를 주어진 경우 위치 X에서 Y보다 같거나 큰 위치로 최소 점프 횟수를 반환한다.

For example, given:

  X = 10
  Y = 85
  D = 30
  
개구리는 다음과 같이 위치하기 때문에 기능은 3을 return해야 한다.

첫 번째 점프 후 위치 10 + 30 = 40
두 번째 점프 후 위치 10 + 30 + 30 = 70
세 번째 점프 후 위치 10 + 30 + 30 + 30 = 100
다음과 같은 가정에 대해 효율적인 알고리즘을 작성하십시오.

X, Y and D are integers within the range [1..1,000,000,000];
X ≤ Y.


```python
# you can write to stdout for debugging purposes, e.g.
# print("this is a debug message")

def solution(X, Y, D):
    # write your code in Python 3.6
    cnt = 0
    while True:
        X = X + D
        cnt = cnt + 1
        if X>=Y:
            break;
        
    return cnt

solution(10, 85, 30)
```

# 풀이 후기

위와 같은 방법으로 풀었더니 퍼포먼스가 0이 나왔다.
while True가 문제인거 같아 카운트를 계속 세는 방법 말고 다른 방법을 찾아봐야겠다.


# 다른 풀이

반복문을 써서 복잡도를 올리는 것보다는 수식으로 바꿔 간단히 하였음.

```python
def solution(X, Y, D):
    # write your code in Python 3.6
    Z = Y - X
    if Z % D == 0:
        return Z // D
    else:
        return Z // D + 1
```
