X, Y and D are integers within the range [1..1,000,000,000];

```python
# you can write to stdout for debugging purposes, e.g.
# print("this is a debug message")

def solution(X, Y, D):
    # write your code in Python 3.6
    cnt = 0
    while True:
        X = X + D
        cnt = cnt + 1
        if X>=Y:
            break;
        
    return cnt

solution(10, 85, 30)
```

```python
def solution(X, Y, D):
    # write your code in Python 3.6
    Z = Y - X
    if Z % D == 0:
        return Z // D
    else:
        return Z // D + 1
```
