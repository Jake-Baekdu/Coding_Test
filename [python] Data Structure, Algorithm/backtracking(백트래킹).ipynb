```python
def is_available(candidate, current_col):
    current_row = len(candidate) # candidate  0 3 
    for queen_row in range(current_row):
        print("current_row", current_row, queen_row) # 1
        print("current_col", current_col) # 0 1 2

        if candidate[queen_row] == current_col or abs(candidate[queen_row] - current_col) == current_row - queen_row:
            return False
    return True


def DFS(N, current_row, current_candidate, final_result):
    if current_row == N: #  row( )  
        final_result.append(current_candidate[:])
        print(final_result)
        return
    
    for candidate_col in range(N): # ex> N=4     
        if is_available(current_candidate, candidate_col):
            current_candidate.append(candidate_col)
            DFS(N, current_row + 1, current_candidate, final_result)
            current_candidate.pop()

def solve_n_queens(N):
    final_result = []
    DFS(N, 0, [], final_result)
    return final_result
```

```python
def nqueen(sol, n):
    global count
    if len(sol) == n: #  (sol)  n , count 
        count += 1
        return 0
    candidate = list(range(n)) # 0 n-1   .
    for i in range(len(sol)):
        if sol[i] in candidate: #     
            candidate.remove(sol[i]) #     
        distance = len(sol) - i
        if sol[i] + distance in candidate: #   (+)   
            candidate.remove(sol[i] + distance) #      
        if sol[i] - distance in candidate: #   (-)   
            candidate.remove(sol[i] - distance) #      
    if candidate != []:
        for i in candidate:
            sol.append(i) #     i+1 
            nqueen(sol, n) #    
    else:
        return 0
count = 0
num = int(input())
for i in range(num): #    
    nqueen([i], num)
print(count)

```
