```python
#틀린 코드
import heapq

n, m, k = map(int, input().split())
List = [[] for _ in range(n+1)]
#k번까지 포장할 수 있으니 k+1만큼 이차원 배열 생성
dist = [[float('inf')]*(k+1) for _ in range(n+1)]
#양방향으로 노드간 거리 생성
for _ in range(m):
    s, e, w = map(int, input().split())
    List[s].append((w, e))
    List[e].append((w, s))

hq = []
# 1에서 1로 가는데 0번 포장했을 경우 = 0
for i in range(k+1):
    dist[1][i] = 0
heapq.heappush(hq, (0, 0, 1))
while hq:
    #가중치, 포장 횟수, 노드
    w, k_cnt, n = heapq.heappop(hq)
    if dist[n][k_cnt] < w:
        continue
    
    #노드가 이어진 곳들 탐색
    for tw, tn in List[n]:
        if k_cnt < k:
            if dist[tn][k_cnt+1] > w:
                dist[tn][k_cnt+1] = w
                heapq.heappush(hq, (w, k_cnt+1, tn))
            
        cost = w + tw
        if dist[tn][k_cnt] > cost:
            dist[tn][k_cnt] = cost
            heapq.heappush(hq, (cost, k_cnt, tn))
print(min(dist[n]))
#######################################################################
#맞은코드
import heapq

n, m, k = map(int, input().split())
List = [[] for _ in range(n+1)]
#k번까지 포장할 수 있으니 k+1만큼 이차원 배열 생성
dist = [[float('inf')]*(k+1) for _ in range(n+1)]
#양방향으로 노드간 거리 생성
for _ in range(m):
    s, e, w = map(int, input().split())
    List[s].append((w, e))
    List[e].append((w, s))

#여기를 함수화 했더니 맞음
def dj(start):
    hq = []
    # 1에서 1로 가는데 0번 포장했을 경우 = 0
    for i in range(k+1):
        dist[start][i] = 0
    heapq.heappush(hq, (0, 0, start))
    while hq:
        #가중치, 포장 횟수, 노드
        w, k_cnt, n = heapq.heappop(hq)
        if dist[n][k_cnt] < w:
            continue
        
        #노드가 이어진 곳들 탐색
        for tw, tn in List[n]:
        	#포장한 횟수가 k보다 작을 경우
            if k_cnt < k:
            	# 지금 가중치(w)가 dist에 들어있는 값보다 작을 경우 dist갱신
                if dist[tn][k_cnt+1] > w:
                    dist[tn][k_cnt+1] = w
                    #tn까지의 가중치(tw)를 포장했으니 heap에 k_cnt + 1로 push 
                    heapq.heappush(hq, (w, k_cnt+1, tn))
            #포장하지 않고 tn까지 가는 경우
            cost = w + tw
            if dist[tn][k_cnt] > cost:
                dist[tn][k_cnt] = cost
                heapq.heappush(hq, (cost, k_cnt, tn))
dj(1)
print(min(dist[n]))
```
