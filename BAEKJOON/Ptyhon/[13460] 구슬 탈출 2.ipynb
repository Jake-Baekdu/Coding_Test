```python
from collections import deque
import sys


N, M = map(int, input().split())
board = [list(input()) for _ in range(N)]
located_blue = []
located_red = []
located_blank = []

for i in range(N):
    for j in range(M):
        if board[i][j] == 'B':
            located_blue = [i, j]
        elif board[i][j] == "R":
            located_red = [i, j]
        elif board[i][j] == "O":
            located_blank = [i, j]


def bfs():
    dq = deque()
    dq.append((located_red, located_blue, 0))
    
    move_xy = [[1,0],[0,1],[-1,0],[0,-1]]
    
    while dq:
        
        red, blue, cnt = dq.popleft()
        
        if cnt >= 10:
            break
        
        for x, y in move_xy:
            new_cnt = cnt + 1
            redPos = red
            bluePos = blue
            redflag = False
            blueflag = False
            board[red[0]][red[1]] = 'R'
            board[blue[0]][blue[1]] = 'B'
            
            while True:
                new_red = [redPos[0]+x, redPos[1]+y]
                new_blue = [bluePos[0]+x, bluePos[1]+y]
                #     
                if new_red == located_blank:
                    redflag = True
                    board[redPos[0]][redPos[1]] = '.'
                    redPos = new_red

                #     
                if new_blue == located_blank:
                    blueflag = True
                    break
                
                #       
                #   (, ) ,       
                if (board[new_red[0]][new_red[1]] != '.' or redflag) and board[new_blue[0]][new_blue[1]] != '.':
                    break
                
                if board[new_red[0]][new_red[1]] == '.' and not redflag:
                    board[redPos[0]][redPos[1]] = '.'
                    redPos = new_red
                    board[redPos[0]][redPos[1]] = 'R'

                if board[new_blue[0]][new_blue[1]] == '.':
                    board[bluePos[0]][bluePos[1]] = '.'
                    bluePos = new_blue
                    board[bluePos[0]][bluePos[1]] = 'B'
                    
            if redflag and not blueflag:
                return new_cnt
                
            if blueflag == False :
                dq.append((redPos, bluePos, new_cnt))
            
            if board[redPos[0]][redPos[1]] == 'R':
                board[redPos[0]][redPos[1]] = '.'
            
            if board[bluePos[0]][bluePos[1]] == 'B':
                board[bluePos[0]][bluePos[1]] = '.'    
    
    return -1        

print(bfs())

```
