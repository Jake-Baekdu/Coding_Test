```python
from collections import deque

N, M = map(int, input().split())
board = [list(input()) for _ in range(N)]

# 파란, 빨간 구슬, 구멍 위치 저장할 변수
located_blue = []
located_red = []
located_blank = []

for i in range(N):
    for j in range(M):
        if board[i][j] == 'B':
            located_blue = [i, j]
        elif board[i][j] == "R":
            located_red = [i, j]
        elif board[i][j] == "O":
            located_blank = [i, j]


def bfs():
    dq = deque()
    dq.append((located_red, located_blue, 0))
    
    # 상하좌우 움직임
    move_xy = [[1,0],[0,1],[-1,0],[0,-1]]
    
    # queue가 빌 때까지
    while dq:
        
        red, blue, cnt = dq.popleft()
        
        # cnt >= 10 이상으로 한 이유는 cnt == 10이면 for문으로 들어가서 11번이 되기 때문
        if cnt >= 10:
            break
        
        for x, y in move_xy:
        	"""
            !!!여기서 주의!!!
            for문(상하좌우 move)을 돌면서 공이 움직이는 위치를 저장해주는 변수를
            위의 red, blue, cnt를 쓰면 안됨! 왜냐하면 queue에서 꺼낸 구슬의 위치에서
            상하좌우를 모두 확인해야되기 때문.
            new_cnt와 redPos와 bluePos를 선언하여 for문 안에서 움직이는 위치를 저장.
            """
            new_cnt = cnt + 1
            redPos = red
            bluePos = blue
            
            #빨간 공이 구멍에 빠졌는지 
            redflag = False
            #파란 공이 구멍에 빠졌는지
            blueflag = False
            
            #for문을 돌면서 공의 위치가 바꼈을 수도 있으니 다시 초기화
            board[red[0]][red[1]] = 'R'
            board[blue[0]][blue[1]] = 'B'
            
            #중력 때문에 끝까지 가야되니 장애물이나 구멍이 있을 때까지 구슬 이동
            while True:
            
            	#우선 한칸 움직인 빨간 구슬의 위치
                new_red = [redPos[0]+x, redPos[1]+y]
                
                #우선 한칸 움직인 파란 구슬의 위치
                new_blue = [bluePos[0]+x, bluePos[1]+y]
                
                # 빨간 구슬이 구멍에 들어가는 경우
                # break를 하지 않는 이유는 파란 구슬이 빠지는지 봐야되기 때문
                if new_red == located_blank:
                    redflag = True
                    board[redPos[0]][redPos[1]] = '.'
                    redPos = new_red

                # 파란 구슬이 구멍에 들어가는 경우
                if new_blue == located_blank:
                    blueflag = True
                    break
                
                # 두 구슬 모두 움직일 수 없는 경우
                # 파란 구슬이 장애물(벽, 빨간구슬)에 막히고, 빨간 구슬도 장애물에 막히거나 구멍에 빠진 경우
                if (board[new_red[0]][new_red[1]] != '.' or redflag) and board[new_blue[0]][new_blue[1]] != '.':
                    break
                
                # 한칸 움직인 빨간 구슬 위치가 이동 가능한 곳이고 빨간 구슬이 구멍에 빠지지 않았을 때
                if board[new_red[0]][new_red[1]] == '.' and not redflag:
                    board[redPos[0]][redPos[1]] = '.'
                    redPos = new_red
                    board[redPos[0]][redPos[1]] = 'R'
				
                # 한칸 움직인 파란 구슬 위치가 이동 가능한 곳일 때
                if board[new_blue[0]][new_blue[1]] == '.':
                    board[bluePos[0]][bluePos[1]] = '.'
                    bluePos = new_blue
                    board[bluePos[0]][bluePos[1]] = 'B'
            
            # 빨간 구슬이 구멍에 빠지고 파란 구슬이 빠지지 않았을 때
            if redflag and not blueflag:
                return new_cnt
            
            # 파란 구슬이 빠지지 않았을 때만 큐에 삽입. 빨간 구슬만 빠졌다면 위에서 return됨
            if blueflag == False :
                dq.append((redPos, bluePos, new_cnt))
            
            # for문을 돌면서 이동한 빨간구슬, 파란구슬 의 위치는 다시 초기화
            if board[redPos[0]][redPos[1]] == 'R':
                board[redPos[0]][redPos[1]] = '.'
            if board[bluePos[0]][bluePos[1]] == 'B':
                board[bluePos[0]][bluePos[1]] = '.'    
    
    #10번이 넘거나 구할 수 없다면 -1 return
    return -1        

print(bfs())
```
