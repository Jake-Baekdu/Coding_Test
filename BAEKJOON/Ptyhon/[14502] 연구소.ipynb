```python
from collections import deque
import copy

x, y = map(int, input().split())
board = [[] for _ in range(x)]
dx = [0, 0, 1, -1]
dy = [1, -1, 0, 0]
ans = 0

for i in range(x):
    board[i] = list(map(int, input().split()))
    
def bfs():
    dq = deque()
    copy_board = copy.deepcopy(board)
    
    for i in range(x):
        for j in range(y):
            if copy_board[i][j] == 2:
                dq.append((i,j))
    
    while dq:
        qx, qy = dq.popleft()
        
        for i in range(4):
            nx = qx + dx[i]
            ny = qy + dy[i]

            if 0 <= nx < x and 0 <= ny < y:
                if copy_board[nx][ny] == 0:
                    copy_board[nx][ny] = 2
                    dq.append((nx, ny))
        
    cnt = 0
    
    for i in range(x):
        cnt += copy_board[i].count(0)
    global ans
    ans = max(ans, cnt)


def makeWall(cnt):
    if cnt == 3:
        bfs()
        return
    
    for i in range(x):
        for j in range(y):
            if board[i][j] == 0:
                board[i][j] = 1
                makeWall(cnt+1)
                board[i][j] = 0
                

makeWall(0)
print(ans)
```
