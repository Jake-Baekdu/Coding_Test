```python
N = int(input())
arr = list(map(int, input().split()))
```

```python
def make_max(arr):
    pl= 0
    pr = 1
    result = 0
    while(pr < len(arr)):
        result += abs(arr[pl] - arr[pr])
        pr += 1
        pl += 1
    return result
```

```python
from itertools import permutations
pe = list(permutations(arr))
maxx = 0
for i in pe:
    temp = make_max(i)
    print(temp)
    if maxx < temp:
        maxx = temp
print(maxx)
```
